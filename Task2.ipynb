Let's take a closer look at our data!

```python
import pandas as pd

import matplotlib
%matplotlib inline
%config InlineBackend.figure_format = 'svg'  #to get high quality images in jupyter notebook
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import seaborn as sns
```

```python
df=pd.read_csv("Data/data_processed.csv", index_col=0)  # index_col specify whether you use the first column as index column or not.
```

```python
df.Region.value_counts()  # .value_counts counts the how many times each unique category exist in a column
```

```python
df.columns.tolist()
```

```python
features=['Channel',
          'Region',
          'Fresh',
          'Milk',
          'Grocery',
          'Frozen',
          'Detergents_Paper',
          'Delicatessen']
```

```python
fig, axes=plt.subplots(2, 4, figsize=(7, 4)) # .subplots provides writing many graphs into one plot. 2 and 4 specify the number of the rows and columns in the fig plot respectively.

for feature, ax in zip(features, axes.ravel()):      # zip provides us to iterate two values in one loop. .ravel() provides us to write multiple figures in a single figure
    if (feature=="Channel") | (feature=="Region"):   # | stands for "or"
        df[feature].value_counts().plot.bar(ax=ax)
    else:
        ax.hist(df[feature], bins=100)
    ax.set_title(feature)
plt.tight_layout()
        
```
